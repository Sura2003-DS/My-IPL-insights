```python
# Total runs scored off the bat
total_runs = df['runs_of_bat'].sum()
print("Total runs scored (off the bat):", total_runs)
```

```python
Top_scorer=df.groupby("striker")["runs_of_bat"].sum().sort_values(ascending=False).head(10)
print("TOP 10 Run scorers:")
Top_scorer
```

```python
wicket_types = ['bowled', 'caught', 'lbw', 'stumped', 'hit wicket']
wickets = df[df['wicket_type'].isin(wicket_types)]
top_bowlers = wickets['bowler'].value_counts().head(10)
print("🎯 Top 10 Wicket Takers:")
top_bowlers
```

```python
valid_wickets=['bowled','caught','lbw','stumped','hit wicket']
wickets_df=df[df['wicket_type'].isin(valid_wickets)]
team_wickets=wickets_df['bowling_team'].value_counts().head(10)
print("Top wickets taken by each team:")
print(team_wickets)
```

```python
sixes = df[df['runs_of_bat'] == 6]['striker'].value_counts().head(10)
print("💥 Top 10 Six Hitters:")
sixes
```

```python
fours_df=df[df['runs_of_bat']==4]
top_four_hitters=fours_df['striker'].value_counts().head(10)
print("Top 10 Four Hitters:")
top_four_hitters
```

```python
team_runs=df.groupby('batting_team')['runs_of_bat'].sum().sort_values(ascending=False)
print("Total runs scored by each team:")
team_runs
```

```python
df['total_extras']=df[['wide','noballs','legbyes','byes']].sum(axis=1)
extras_by_team=df.groupby('bowling_team')['total_extras'].sum().sort_values(ascending=False)
print("Extras conceded by each team in IPL 2023:")
extras_by_team
```

```python
# Filter deliveries where no runs were scored off the bat
dot_balls = df[df['runs_of_bat'] == 0]

# Count how many dot balls each bowler delivered
top_dot_bowlers = dot_balls['bowler'].value_counts().head(10)

print("🎯 Top 10 Dot Ball Bowlers:")
print(top_dot_bowlers)
```

```python
# Total runs scored by each batsman
runs_by_batsman = df.groupby('striker')['runs_of_bat'].sum()

# Balls faced by each batsman (excluding extras like wides)
balls_faced = df[df['wide'] == 0].groupby('striker').size()

# Filter batsmen with at least 100 balls faced
qualified_batsmen = balls_faced[balls_faced >= 100]

# Calculate strike rate = (runs / balls) * 100
strike_rate = (runs_by_batsman[qualified_batsmen.index] / qualified_batsmen) * 100

# Get top 10
top_strikers = strike_rate.sort_values(ascending=False).head(10)

print("⚡ Top 10 Batsmen with Highest Strike Rates (min 100 balls faced):")
print(top_strikers)
```

```python
# Dot ball = no run from bat and not an extra (wide, no-ball, etc.)
dot_balls = df[(df['runs_of_bat'] == 0) &
               (df['extras'] == 0)]

# Count dot balls bowled by each bowler
dot_ball_count = dot_balls['bowler'].value_counts().head(10)

print("🧊 Top 10 Bowlers with Most Dot Balls:")
print(dot_ball_count)
```

```python
# Filter for 4s and 6s
boundaries = df[df['runs_of_bat'].isin([4, 6])]

# Count boundaries conceded by each bowler
boundary_conceded = boundaries['bowler'].value_counts().head(10)

print("🚨 Top 10 Bowlers Who Conceded the Most Boundaries:")
print(boundary_conceded)
```

```python
import matplotlib.pyplot as plt

# Dot balls (runs_of_bat == 0)
dot_balls = df[df['runs_of_bat'] == 0]

# Total balls bowled by each bowler
total_balls = df.groupby('bowler').size()

# Dot ball count
dot_ball_count = dot_balls['bowler'].value_counts()

# Combine into a DataFrame
dot_stats = pd.DataFrame({
    'dot_balls': dot_ball_count,
    'total_balls': total_balls
})

# Drop NaNs and calculate percentage
dot_stats.dropna(inplace=True)
dot_stats['dot_percent'] = (dot_stats['dot_balls'] / dot_stats['total_balls']) * 100

# Sort and get top 10
top_dot_bowlers = dot_stats.sort_values(by='dot_balls', ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
plt.barh(top_dot_bowlers.index[::-1], top_dot_bowlers['dot_balls'][::-1], color='brown')
plt.xlabel("Dot Balls")
plt.title("Top 10 Bowlers by Dot balls per match")
plt.tight_layout()
plt.show()

# Print with percentages
print(top_dot_bowlers[['dot_balls', 'total_balls', 'dot_percent']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count missing values
missing_counts = df[['wicket_type', 'player_dismissed', 'fielder']].isnull().sum().reset_index()
missing_counts.columns = ['Column', 'Missing Values']

# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(data=missing_counts, x='Column', y='Missing Values', hue='Column', palette='Set2', legend=False)
plt.title("Missing Values in Wicket-Related Columns", fontsize=14)
plt.ylabel("Number of Missing Values")
plt.xlabel("Column Name")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count missing values
missing_counts = df[['wicket_type', 'player_dismissed', 'fielder']].isnull().sum()

# Create horizontal bar plot
plt.figure(figsize=(8, 4))
missing_counts.sort_values().plot(kind='barh', color='green', edgecolor='black')

plt.title("Missing Values in Wicket-related Columns", fontsize=14)
plt.xlabel("Number of Missing Values")
plt.ylabel("Column Name")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
